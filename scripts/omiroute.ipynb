```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from datetime import datetime
import os

# =========================================================
# 1. Spark Session
# =========================================================
spark = SparkSession.builder \
    .appName("VehicleRegistryPipeline") \
    .getOrCreate()

# =========================================================
# 2. PATHS
# =========================================================
BASE_PATH = "/Users/harshvardhan/Downloads/bootcamp-project/bootcamp-project"

RAW_PATH = f"{BASE_PATH}/bronze/vehicle_registry/raw_files/"
CORRUPT_PATH = f"{BASE_PATH}/bronze/vehicle_registry/corrupted_data/"
LOG_PATH = f"{BASE_PATH}/bronze/vehicle_registry/logs/"
SILVER_PATH = f"{BASE_PATH}/silver/vehicle_registry_clean/"
PROCESSED_FILE_LOG = f"{LOG_PATH}/processed_files.csv"

VIN_LEN = 8
CURRENT_YEAR = datetime.now().year

# =========================================================
# 3. GET FILE LIST
# =========================================================
all_files = [f for f in os.listdir(RAW_PATH) if f.endswith(".csv")]

# =========================================================
# 4. READ PROCESSED FILE LOG
# =========================================================
if os.path.exists(PROCESSED_FILE_LOG):
    processed_df = spark.read.option("header", True).csv(PROCESSED_FILE_LOG)
    processed_files = [row["file_name"] for row in processed_df.collect()]
else:
    processed_files = []

# =========================================================
# 5. FILTER NEW FILES
# =========================================================
new_files = [f for f in all_files if f not in processed_files]

if not new_files:
    print("✅ No new files to process")
    exit()

print("📂 New files:", new_files)

new_file_paths = [RAW_PATH + f for f in new_files]

# =========================================================
# 6. READ DATA
# =========================================================
df_raw = spark.read \
    .option("header", True) \
    .option("mode", "PERMISSIVE") \
    .csv(new_file_paths)

# Track source file
df_raw = df_raw.withColumn("source_file", F.input_file_name())

print("Total Raw Rows:", df_raw.count())

# =========================================================
# 7. STANDARDIZATION
# =========================================================
df = df_raw.select(
    F.trim("vin").alias("vin"),
    F.trim("model").alias("model"),
    F.trim("mfg_year").alias("mfg_year"),
    F.trim("fuel_type").alias("fuel_type"),
    F.trim("base_kmpl").alias("base_kmpl"),
    "source_file"
)

df = df.withColumn("vin", F.upper("vin"))
df = df.withColumn("model", F.initcap("model"))
df = df.withColumn("fuel_type", F.upper("fuel_type"))

# =========================================================
# 8. TYPE CASTING
# =========================================================
df = df.withColumn(
    "mfg_year_int",
    F.regexp_extract("mfg_year", r"(-?\d+)", 1).cast("int")
)

df = df.withColumn(
    "base_kmpl_double",
    F.regexp_extract("base_kmpl", r"(\d+(\.\d+)?)", 1).cast("double")
)

# =========================================================
# 9. VALIDATION RULES
# =========================================================
valid_fuels = ["DIESEL", "PETROL", "LNG", "CNG", "ELECTRIC", "HYBRID"]

df = df.withColumn("vin_error",
    F.when(F.col("vin").isNull() | (F.length("vin") == 0), "VIN_MISSING")
     .when(F.length("vin") != VIN_LEN, "VIN_INVALID_LENGTH")
     .when(~F.col("vin").rlike("^[A-Z0-9]+$"), "VIN_INVALID_CHAR")
)

df = df.withColumn("model_error",
    F.when(F.col("model").isNull() | (F.length("model") == 0), "MODEL_MISSING")
)

df = df.withColumn("year_error",
    F.when(F.col("mfg_year_int").isNull(), "YEAR_INVALID")
     .when(F.col("mfg_year_int") < 1980, "YEAR_TOO_OLD")
     .when(F.col("mfg_year_int") > CURRENT_YEAR, "YEAR_FUTURE")
)

df = df.withColumn("fuel_error",
    F.when(~F.col("fuel_type").isin(valid_fuels), "FUEL_INVALID")
)

df = df.withColumn("kmpl_error",
    F.when(F.col("base_kmpl_double").isNull(), "KMPL_INVALID")
     .when(F.col("base_kmpl_double") <= 0, "KMPL_NEGATIVE_OR_ZERO")
     .when(F.col("base_kmpl_double") > 60, "KMPL_TOO_HIGH")
)

# =========================================================
# 10. MERGE ERRORS
# =========================================================
df = df.withColumn(
    "reject_reason",
    F.concat_ws("|",
        "vin_error", "model_error", "year_error", "fuel_error", "kmpl_error"
    )
)

df = df.withColumn(
    "reject_reason",
    F.when(F.col("reject_reason") == "", None).otherwise(F.col("reject_reason"))
)

# =========================================================
# 11. SPLIT DATA
# =========================================================
clean_df = df.filter(F.col("reject_reason").isNull())

reject_df = df.filter(F.col("reject_reason").isNotNull())

# =========================================================
# 12. FINAL CLEAN DATA
# =========================================================
clean_df = clean_df.select(
    "vin",
    "model",
    F.col("mfg_year_int").alias("mfg_year"),
    "fuel_type",
    F.round("base_kmpl_double", 2).alias("base_kmpl"),
    "source_file"
).dropDuplicates(["vin"])

# =========================================================
# 13. WRITE OUTPUTS
# =========================================================

# ✅ Silver Layer
clean_df.write.mode("append").parquet(SILVER_PATH)

# ❌ Corrupted Data
reject_df.write.mode("append").option("header", True).csv(CORRUPT_PATH)

# =========================================================
# 14. LOGGING METRICS
# =========================================================
total = df.count()
valid = clean_df.count()
rejected = reject_df.count()

log_data = [(total, valid, rejected, str(datetime.now()))]

log_df = spark.createDataFrame(log_data, ["total", "valid", "rejected", "timestamp"])

log_df.write.mode("append").option("header", True).csv(LOG_PATH)

# =========================================================
# 15. UPDATE PROCESSED FILES LOG
# =========================================================
processed_data = [(f, str(datetime.now())) for f in new_files]

processed_df = spark.createDataFrame(processed_data, ["file_name", "processed_time"])

if os.path.exists(PROCESSED_FILE_LOG):
    processed_df.write.mode("append").option("header", True).csv(PROCESSED_FILE_LOG)
else:
    processed_df.write.mode("overwrite").option("header", True).csv(PROCESSED_FILE_LOG)

# =========================================================
# DONE
# =========================================================
print("✅ Pipeline Completed")
print(f"Total: {total}, Valid: {valid}, Rejected: {rejected}")
```
